## 2. Important libraries for physics:

## 3. Vector generation:

```python
# Vectors for the X axis and Y axis

x = np.arange(-2*np.pi, 2*np.pi, 0.01)

y = np.cos(x)

print(x.shape, y.shape)
```

## 3. 1D plots:

```python
# For the plot we use Matplotlib

plt.figure(figsize = (5,4))

plt.plot(x, y, color = "magenta", linestyle = ":", label = r"$y = \cos(x)$")

plt.xlabel(r"$x$")
plt.ylabel(r"$y$")

plt.legend(loc=1)

plt.show()
```

## 4. 2D Meshgrids:

```python
# We need to create a 2D grid, based on 1D vectors

# Let's create the 1D vectors
x = np.linspace(-10., 10., 1000)
y = np.linspace(-10., 10., 1000)

# Now we need the 2D grid
x_2d, y_2d = np.meshgrid(x, y)

print(x.shape, x_2d.shape)
print(y.shape, y_2d.shape)
```

## 4. 2D plots:

Surface plots:

$$z = \sin\left(\sqrt{x^2+y^2}\right)$$

```python
# Let's define a python function

def surface_1(x_2d, y_2d):
  """
  This is a function to create a 2D surface.
  Inputs: x_2D, y_2D
  Outputs: z_2D
  """

  z_2d = np.sin(x_2d) + np.cos(y_2d)

  return z_2d
```

```python
# Call the surface function

z_2d = surface_1(x_2d, y_2d)

print(z_2d.shape)
```

```python
from mpl_toolkits.mplot3d import Axes3D
```

```python
# Plotting

fig = plt.figure(figsize=(6,5))
ax = fig.add_subplot(111, projection='3d')

ax.plot_surface(x_2d, y_2d, z_2d, cmap = "viridis")

ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')

plt.show()
```
