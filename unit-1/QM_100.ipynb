```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
def psi_function(a, n, x):
    
    psi_x = np.sqrt(2/a)*np.sin(n*np.pi*x/a)
    
    return psi_x
```

```python
a = 2*np.pi

#n = 1

x = np.arange(0, 2*np.pi, 0.01)
```

```python
y1 = psi_function(a, 1, x)

y2 = psi_function(a, 2, x)

y3 = psi_function(a, 3, x)

y4 = psi_function(a, 4, x)
```

```python
%matplotlib inline

fig, ax = plt.subplots(figsize=(10,10))
ax.plot(x,y1, label = r'$\psi_1$')

ax.plot(x,y2)

ax.plot(x,y3)

ax.plot(x,y4)

ax.legend()
```

```python
def psi_function(a, n, x):
    if(x <= 0):
        return 0
    elif(0 <= x <= a):
        return np.sqrt(2/a)*np.sin(n*np.pi*x/a)
    else:
        return 0
```

```python
x_general = np.arange(-2*np.pi, +4*np.pi, 0.01)
y_general = []

for i in range(len(x_general)):
    y_general.append(psi_function(a, 4, x_general[i]))

#print(y)
```

```python
plt.plot(x_general, y_general, c = 'red')
```
