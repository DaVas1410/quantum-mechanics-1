# Summary:

## A) Free particles:

$V(x) = 0$

### Solution (wave packets):

1. Envelope -> Gaussian function that moves at the group velocity ($v_{group} = \frac{d\omega}{dk}$)


2. Ripples -> Harmonic functions that move at th phase velocity ($v_{phase}=\frac{\omega}{k}$).

### Example:

$\psi(x) = 5\sin(2\pi\alpha x)\,e^{-x^2/2}$

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Vector for the x axis
#x = np.linspace(-3*np.pi, +3*np.pi, 1000)

x = np.arange(-3*np.pi, +3*np.pi, 0.01)


# psi function

def psi_free_particle(x, alpha):

  y_axis = 5*np.sin(2*np.pi*alpha*x)*np.exp(-x**2/2)

  return y_axis

# Vector for the psi function (y axis)

y_1 = psi_free_particle(x, 1) # for alpha =1
y_2 = psi_free_particle(x, 2) # for alpha =2
y_3 = psi_free_particle(x, 3) # for alpha =3
y_4 = psi_free_particle(x, 4) # for alpha =4
```

```python
# Check:

print(x.shape, y_1.shape, y_2.shape)
```

```python
# Plotting

plt.figure(figsize=(6,5))

plt.plot(x, y_1, linestyle = "-", label=r"$\alpha = 1$")
plt.plot(x, y_2, linestyle = ":", label=r"$\alpha = 2$")

plt.legend()
plt.show()
```

```python
# Multi-panel figure, ax = axes

fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize = (8,4))

ax1.plot(x, y_1, linestyle = "-", color = 'red',  label=r"$\alpha = 1$")
ax2.plot(x, y_2, linestyle = "-", color = 'magenta', label=r"$\alpha = 2$")
ax3.plot(x, y_3, linestyle = "-", color = 'green', label=r"$\alpha = 3$")
ax4.plot(x, y_4, linestyle = "-", label=r"$\alpha = 4$")

ax1.legend()
ax2.legend()
ax3.legend()
ax4.legend()
plt.show()
```

## 2D Gaussian Wave Packets:

Consider the following 2D wavefunction:

$\psi(x, y) = \sin(\alpha x)\,e^{-\lambda(x^2 + y^2)}$
 
Create a 2D representation.

```python
# Function for the 2D Gaussian wavepacket:

def psi2D_free_particle(l, a, x_2D, y_2D):
  # l = lambda in the eq. above

  f_xy = np.sin(a*x_2D)*np.exp(-l*(x_2D**2 + y_2D**2))

  return f_xy
```

```python
# Axes -> 1 D vectors
x_axis = np.arange(-2*np.pi, +2*np.pi, 0.05)
y_axis = np.arange(-2*np.pi, +2*np.pi, 0.05)

# Mesh grid: 2D matrices with the coordinates
X, Y = np.meshgrid(x_axis, y_axis)

print(X.shape, Y.shape)
```

```python
# Call the function:
l_factor = 0.5

# Alpha
alpha = 4.

# Surface:
psi_xy = psi2D_free_particle(l_factor, alpha, X, Y)
```

```python
# 2D Plotting

plt.figure(figsize=(6, 5))

Z1 = plt.pcolor(X, Y, psi_xy)

plt.colorbar(Z1)

plt.show()
```

```python
# Surface Plotting
fig, ax = plt.subplots(figsize=(8,5), subplot_kw={"projection": "3d"})

# Plot the surface.
surf = ax.plot_surface(X, Y, psi_xy, cmap = 'viridis', antialiased=False)
plt.colorbar(surf)

plt.xlabel("x")
plt.ylabel("y")

plt.show()
```
